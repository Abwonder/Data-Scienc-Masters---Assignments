### Question 1

Q1. Create a function which will take a list as an argument and return the product of all the numbers
after creating a flat list.
Use the below-given list as an argument for your function.

list1 = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45,
22, 61, 34)}, [56, 'data science'], 'Machine Learning']

Note: you must extract numeric keys and values of the dictionary also.

```python
## Code to creat flat list and product of it numbers
def product_list_numbers(list_input):
    list2= []   ## newlist to append and work finally with with functions
    for i in list_input:   ## this handles single element of the types listed below
        if type(i) == int or type(i) == str or type(i) == bool or type(i) == float:
            list2.append(i)

        elif type(i) == list or type(i) == tuple or type(i) == set:   # here i will be a set of list--- easy to iterate over
                for j in i:
                    list2.append(j)
            
        elif type(i) == dict:
            for j in i:
                list2.append(j)
                if type(i[j]) == list or type(i[j])==tuple or type(i[j])==set:
                    for m in (i[j]):
                        list2.append(m)
                else:
                    list2.append(i[j])
    list_use_product = []
    for i in list2:
        if type(i) == int or type(i) == float:
            list_use_product.append(i)
    # return list_use_product,  ## this returns just the numbers in the list
    sq_funtion = lambda x,y: x*y
    from functools import reduce
    result = reduce(sq_funtion, list_use_product)
    return result
```

```python
list1 = [1,2,3,4, [44,55,66, True], False, (34,56,78,89,34), {1,2,3,3,2,1}, {1:34, "key2": [55, 67, 78, 89], 4: (45, 22, 61, 34)}, [56, 'data science']]
product_list_numbers(list1)  ## function and the input
```

Q2. Write a python program for encrypting a message sent to you by your friend. The logic of encryption
should be such that, for a the output should be z. For b, the output should be y. For c, the output should
be x respectively. Also, the whitespace should be replaced with a dollar sign. Keep the punctuation
marks unchanged.

Input Sentence: I want to become a Data Scientist.

Encrypt the above input sentence using the program you just created.

Note: Convert the given input sentence into lowercase before encrypting. The final output should be
lowercase.

```python
def encrypr_text(text):
    alphabet_list= list("abcdefghijklmnopqrstuvwxyz")
    main = str(text.lower())
    # count=-1 
    reverse_count = 0
    new_str = ""
    for i in main:
        # print(i)
        # i.isalpha()
        if i.isalpha() == True:  ## this majorly for the alphabets reverse
            # print(i)
            for a_str in alphabet_list:
                reverse_count-=1
                # count+=1
                # print(f"{reverse_count}, {count}")
                if i == a_str:
                    new_str+=alphabet_list[reverse_count]
                    # print(new_str)
                    break
            # print(f"{reverse_count}, {count}") 
            reverse_count=0
            # count=-1

        elif i == " ":
            # print("$")
            new_str+="$"
            # print(new_str)
        else:  ## this will deal with punctuation and those with nu condition in this code
            new_str+=i
    return new_str
```

```python
the_text = "I want to become a Data Scientist."
encrypr_text(the_text)
```
