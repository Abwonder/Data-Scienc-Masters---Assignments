Question 5

* To to run set of code for undefine number of times until defined conditions are met.
* When running a loop that does not have predefined cycles that will continue to run until a particular conditioin is met!! and when loop is not based on predefined iteration or range.

I give example below:

```python
# stock inventory that needs to alert business when their stock gets to a particular level

shoe_stock = int(input(f"Enter number of shoe stocked:  \n"))

count = shoe_stock

while count > 50:
    number_bought = int(input("How many shoes bought: "))
    count = count - number_bought
print(f"Stock is now below 50 and presently {count}")
print(f"Please restock shoe to conitue selling!!!")
```
