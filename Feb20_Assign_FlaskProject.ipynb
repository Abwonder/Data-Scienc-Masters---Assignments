### Question 1

Q1. Explain GET and POST methods.

GET method ==> is mainly used at the client (Browser) side to send a request to a specified server to get certain data or resources, and it is the most use HTTP method. The working principle goes thus: the request parameter of GET is appended to the URL (content visible on the url address), not secured and only use to view resources but not to change it.

POST method ==> is mainly used at the client (Browser) side to send data to a Specified server in order to create or rewrite a particular resource/data. This data sent to the server is stored in the request body of the HTTP request. Thus, this method is useful when resources are to be changed or updated.
