### Question 5
Q5. How are DataFrame.size() and DataFrame.shape() different?

DataFrame.size() and DataFrame.shape() different in the information they output when used, although both of them are Pandas DataFrame attributes: DataFrame.size() returns the total number of elements in the DataFrame which is the multiplication of the rows and columns i.e 2 by 3 (6 )while DataFrame.shape returns a tuple that contains the row and column numbers e.g  (2,3).



```python
## An example of the above
import pandas as pd

# Create a sample DataFrame
df = pd.DataFrame({'A': [1, 2, 3], 'B': [4, 5, 6], 'C': [7, 8, 9], 'D': [10, 11, 12]})

# Get the size and shape of the DataFrame
print('DataFrame size:', df.size)
print('DataFrame shape:', df.shape)
```

### Question 10
Q10. Given a Pandas DataFrame df with a column 'Sales' and a column 'Date', write a Python function to
create a new column 'MovingAverage' that contains the moving average of the sales for the past 7 days
for each row in the DataFrame. The moving average should be calculated using a window of size 7 and
should include the current day.

```python
import pandas as pd
#datasetframe

def moving_avg_add(df):
### arrange the dates in orderly manner
    df = df.sort_values('Date')

    # Calculate the rolling average with a window of 7 days, including the current day
    rolling_avg = df['Sales'].rolling(window=7, min_periods=1).mean()

    # Add the rolling average as a new column in the DataFrame
    df['MovingAverage'] = rolling_avg
    
    return df

## adding the moving average colunm to the DataFrame
df = moving_avg_add(df)
```

### Question 11
Q11 You have a Pandas DataFrame df with a column 'Date'. Write a Python function that creates a new
column 'Weekday' in the DataFrame. The 'Weekday' column should contain the weekday name (e.g.
Monday, Tuesday) corresponding to each date in the 'Date' column.
For example, if df contains the following values:
Date
0 2023-01-01
1 2023-01-02
2 2023-01-03
3 2023-01-04
4 2023-01-05
Your function should create the following DataFrame:

Date Weekday
0 2023-01-01 Sunday
1 2023-01-02 Monday
2 2023-01-03 Tuesday
3 2023-01-04 Wednesday
4 2023-01-05 Thursday
The function should return the modified DataFrame.

```python
## Will use this example to answer the question above
import pandas as pd
## creating the dataframe
df = pd.DataFrame({"Date":["2023-01-01", "2023-01-02", "2023-01-03", "2023-01-04", "2023-01-05"]})

## creating the function to add weekday column
def weekday_col_add(df):
    ## convert it to datetime format while picking it out as series
    df["Date"] = pd.to_datetime(df["Date"])
    ## creating new column in weekday format
    df["Weekday"] = df["Date"].dt.day_name()
    return df

df = weekday_col_add(df)
df
```
